```python
import keras
from keras import layers, initializers
import numpy as np
import tensorflow
import tensorflow as tf
import glob
from sklearn.preprocessing import MinMaxScaler
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

# Load

```python
files = glob.glob('./datasets/AKWF/*.wav')
print('There are:',len(files), 'files.')
data = np.empty((1,600)) #create empty array with same shape as data will be
print(data.shape)
mms = MinMaxScaler(feature_range = (0,1))
for file in files:
    samplerate, wav_data = wavfile.read(file)
    
    if wav_data.shape == (600,):

        wav_data = wav_data.reshape(wav_data.shape[0],1)
        t = mms.fit_transform(wav_data)
        t = t.reshape(1,t.shape[0])
        t = np.nan_to_num(t)
    else:
        print('Encountered multi-channel wav:', file)
    data = np.append(data,t,axis = 0) #append along correct axis
```

```python
# data = data[-100::]
```

```python
data = data.reshape(data.shape[0],data.shape[-1])
print(data.shape)
```

# Encoder Model

WHEN USING RELU INIZIALIZE WITH SMALL BIAS, OR ELSE POTENTIAL FOR DEAD NEURONS

Initializing with random_uniform also helps with dead neurons

```python
from keras import backend as KB
```

```python
data_shape = data.shape[-1]
encoding_dimension = 2

input_vector = keras.Input(shape = (data_shape))
e1 = layers.Dense(300,activation = 'sigmoid',kernel_initializer = 'random_uniform')(input_vector)
e1 = layers.BatchNormalization()(e1)
# e1 = layers.Dropout(0.5)(e1)
# e1_c = layers.concatenate([input_vector, e1])



e2 = layers.Dense(150,activation = 'sigmoid',kernel_initializer = 'random_uniform')(e1)
e2 = layers.BatchNormalization()(e2)
# e2 = layers.Dropout(0.5)(e2)
# e2_c = layers.concatenate([input_vector, e1, e2])


e3 = layers.Dense(75,activation = 'sigmoid',kernel_initializer = 'random_uniform')(e2)
# e3 = layers.MaxPooling1D(pool_size=2, strides=2, padding='valid')(e3)
e3 = layers.BatchNormalization()(e3)
# e3 = layers.Dropout(0.5)(e3)
# e3_c = layers.concatenate([input_vector, e1, e2, e3])


e4 = layers.Dense(37,activation = 'sigmoid',kernel_initializer = 'random_uniform')(e3)
e4 = layers.BatchNormalization()(e4)
# e4 = layers.Dropout(0.5)(e4)
# e4_c = layers.concatenate([input_vector, e1, e2, e3, e4])

e_out = layers.Flatten()(e4)
e_out = layers.Dense(encoding_dimension,activation = 'sigmoid',kernel_initializer = 'random_uniform')(e4)
# e_out = layers.Reshape((1,e_out.shape[-1]))(e_out)






input_decoder = keras.Input(shape=(encoding_dimension,))
d1 = layers.Dense(37,activation = 'sigmoid',kernel_initializer = 'random_uniform')(input_decoder)
d1 = layers.BatchNormalization()(d1)
print(d1.shape)

d2 = layers.Dense(75,activation = 'sigmoid',kernel_initializer = 'random_uniform')(d1)
d2 = layers.BatchNormalization()(d2)
# d2 = layers.Dropout(0.5)(d2)
# d2_c = layers.concatenate([d1,d2])



d3 = layers.Dense(150,activation = 'sigmoid',kernel_initializer = 'random_uniform')(d2)
d3 = layers.BatchNormalization()(d3)
# d3 = layers.Dropout(0.5)(d3)
# d3_c = layers.concatenate([d1,d2,d3])


d4 = layers.Dense(300,activation = 'sigmoid',kernel_initializer = 'random_uniform')(d3)
d4 = layers.BatchNormalization()(d4)
# d4 = layers.Dropout(0.5)(d4)
# d4_c = layers.concatenate([d1,d2,d3,d4])


d_out = layers.Dense(600, activation='sigmoid',kernel_initializer = 'random_uniform')(d4)
# d_out = layers.Reshape(data_size)(d_out)




encoder = keras.Model(input_vector,e_out ,name = 'encoder')
decoder = keras.Model(input_decoder,d_out,name = 'd')

encoder.summary()
decoder.summary()

output = decoder(encoder(input_vector))
stacked = keras.Model(input_vector,output)


stacked.compile(optimizer = 'adam', loss = 'binary_crossentropy')
stacked.summary()


```

```python
stacked.fit(x = data,y = data, epochs = 5,batch_size = 128, validation_data = (data,data))
```

# Look at Latent Space 

```python
mean = encoder.predict(data)
print(mean.shape)
plt.figure(figsize=(10,10))
plt.scatter(mean[:,0], mean[:,1])
# print(mean)
plt.savefig('latent_space.png')
```

# Generate New Sample From Dimension Space

```python
mean_sample = np.array([np.mean(mean[:,1]),np.mean(mean[:,0])]).reshape(1,2)
random_sample = np.array([np.random.uniform(), np.random.uniform()]).reshape(1,2)
x = 0.1
y = 0.1
coord_sample = np.array([x,y]).reshape(1,2)



x_mean = decoder.predict(mean_sample)
x_random = decoder.predict(random_sample)
x_coord = decoder.predict(coord_sample)

plt.figure(figsize = (10,10))
plt.plot(x_mean[0], label = 'mean')
plt.plot(x_random[0], label = 'random')
plt.plot(x_coord[0], label = 'coord')
plt.legend()
```
