```python
import keras
from keras import layers, initializers
import numpy as np
import tensorflow
import tensorflow as tf
import glob
from sklearn.preprocessing import MinMaxScaler
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

# Load

```python
files = glob.glob('./datasets/AKWF/*.wav')
print('There are:',len(files), 'files.')
data = np.empty((1,600)) #create empty array with same shape as data will be
print(data.shape)
mms = MinMaxScaler(feature_range = (0,1))
for file in files:
    samplerate, wav_data = wavfile.read(file)
    
    if wav_data.shape == (600,):

        wav_data = wav_data.reshape(wav_data.shape[0],1)
        t = mms.fit_transform(wav_data)
        t = t.reshape(1,t.shape[0])
    else:
        print('Encountered multi-channel wav:', file)
    data = np.append(data,t,axis = 0) #append along correct axis
```

```python
# data = data[-2::]
```

```python
data = data.reshape(data.shape[0],data.shape[-1],1)
print(data.shape)
```

```python
data_shape = data.shape[1::]
encoding_dimension = 2

input_vector = keras.Input(shape = (data_shape))
e1 = layers.Conv1D(300,3,activation = 'relu',padding = 'same')(input_vector)
e1 = layers.MaxPooling1D(pool_size = 3)(e1)
e1 = layers.BatchNormalization()(e1)
e1 = layers.Dropout(0.5)(e1)
# e1_c = layers.concatenate([input_vector, e1])



e2 = layers.Conv1D(150,3,activation = 'relu', padding = 'same')(e1)
e2 = layers.MaxPooling1D(pool_size = 2)(e2)
e2 = layers.BatchNormalization()(e2)
e2 = layers.Dropout(0.5)(e2)
# e2_c = layers.concatenate([input_vector, e1, e2])


e3 = layers.Conv1D(75,3,activation = 'relu',padding = 'same')(e2)
e3 = layers.MaxPooling1D(pool_size=2)(e3)
e3 = layers.BatchNormalization()(e3)
e3 = layers.Dropout(0.5)(e3)
# # e3_c = layers.concatenate([input_vector, e1, e2, e3])


e4 = layers.Conv1D(37,3,activation = 'relu',padding = 'same')(e3)
e4 = layers.MaxPooling1D(pool_size = 2)(e4)
e4 = layers.BatchNormalization()(e4)
e4 = layers.Dropout(0.5)(e4)
# e4_c = layers.concatenate([input_vector, e1, e2, e3, e4])

e_out = layers.Flatten()(e4)
e_out = layers.Dense(encoding_dimension,activation = 'sigmoid')(e_out)
# e_out = layers.Reshape((1,e_out.shape[-1]))(e_out)






input_decoder = keras.Input(shape=(encoding_dimension,))
d1 = layers.Dense(25 * 37, activation = 'relu')(input_decoder)
d1 = layers.Reshape((25,37))(d1)


d1 = layers.UpSampling1D(size = 2)(d1)
d1 = layers.Conv1DTranspose(37,3,activation = 'relu', padding = 'same')(d1)
d1 = layers.BatchNormalization()(d1)


d2 = layers.UpSampling1D(size = 2)(d1)
d2 = layers.Conv1DTranspose(75,3,activation = 'relu', padding = 'same')(d2)
d2 = layers.BatchNormalization()(d2)
d2 = layers.Dropout(0.5)(d2)
# d2_c = layers.concatenate([d1,d2])


d3 = layers.UpSampling1D(size = 2)(d2)
d3 = layers.Conv1DTranspose(150,3,activation = 'relu', padding = 'same')(d3)
d3 = layers.BatchNormalization()(d3)
d3 = layers.Dropout(0.5)(d3)
# d3_c = layers.concatenate([d1,d2,d3])

d4 = layers.UpSampling1D(size = 3)(d3)
d4 = layers.Conv1DTranspose(300,3,activation = 'relu', padding = 'same')(d4)
d4 = layers.BatchNormalization()(d4)
d4 = layers.Dropout(0.5)(d4)


d_out = layers.Conv1D(1,3, activation = 'sigmoid', padding = 'same')(d4)
d_out = layers.BatchNormalization()(d_out)



encoder = keras.Model(input_vector,e_out ,name = 'encoder')
decoder = keras.Model(input_decoder,d_out,name = 'd')

encoder.summary()
decoder.summary()

output = decoder(encoder(input_vector))
stacked = keras.Model(input_vector,output)


stacked.compile(optimizer = 'adam', loss = 'mean_squared_error')
stacked.summary()


```

```python
cb = tf.keras.callbacks.EarlyStopping(monitor = 'val_loss', patience = 3)
stacked.fit(x = data,y = data, epochs = 50,batch_size = 128, validation_data = (data,data), callbacks = [cb] )


```

```python
mean = encoder.predict(data)
print(mean.shape)
plt.figure(figsize=(10,10))
plt.scatter(mean[:,0], mean[:,1])
# print(mean)
plt.savefig('latent_space.png')
```

# Generate New Sample From Dimension Space

```python
mean_sample = np.array([np.mean(mean[:,1]),np.mean(mean[:,0])]).reshape(1,2)
random_sample = np.array([np.random.uniform(), np.random.uniform()]).reshape(1,2)
x = 0.1
y = 0.1
coord_sample = np.array([x,y]).reshape(1,2)

scaler = MinMaxScaler(feature_range = (-1,1))

x_mean = decoder.predict(mean_sample)
x_mean = scaler.fit_transform(x_mean[0])


x_random = decoder.predict(random_sample)
x_random = scaler.fit_transform(x_random[0])


x_coord = decoder.predict(coord_sample)
x_coord = scaler.fit_transform(x_coord[0])
plt.figure(figsize = (10,10))
plt.plot(x_mean, label = 'mean')
plt.plot(x_random, label = 'random')
plt.plot(x_coord, label = 'coord')
plt.legend()
```
